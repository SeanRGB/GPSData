```python
import datetime as dt
import ftplib
from ftplib import FTP
```

```python
ftpSite = 'sideshow.jpl.nasa.gov'
path2Files = '/pub/JPL_GPS_Timeseries/repro2011b/post/point/'
testFile = '019B.lat'
```

```python
site = FTP(ftpSite)
site.login()
site.cwd(path2Files)
```

```python
def monthName2Num(monthName):
    return{
        'Jan' : 1,
        'Feb' : 2,
        'Mar' : 3,
        'Apr' : 4,
        'May' : 5, 
        'Jun' : 6,
        'Jul' : 7,
        'Aug' : 8,
        'Sep' : 9,
        'Oct' : 10,
        'Nov' : 11,
        'Dec' : 12
        }[monthName]
```

```python
rawManifest = []

line = site.retrlines('LIST', rawManifest.append)
```

```python
manifest = {}

for index in range(len(rawManifest)):
    lineList = rawManifest[index].split()
    if (lineList[7][2] == ':'):
        manifest[lineList[8]] = [monthName2Num(lineList[5]), int(lineList[6]), 2015]
    else:
        manifest[lineList[8]] = [monthName2Num(lineList[5]), int(lineList[6]), int(lineList[7])]
```
